```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, boxcox
from statsmodels.nonparametric.smoothers_lowess import lowess
from sklearn.preprocessing import PowerTransformer

# Load the data
df = pivot_df.copy()

# Choose specific indicators and years for analysis
chosen_indicators = ["Sales", "Total Liabilities"]  # Example indicators
years = ["2023", "2022"]

# Iterate over each indicator and year
for indicator in chosen_indicators:
    for year in years:
        # Construct column names for seller and buyer values for the chosen year
        seller_col = f"SELL-SIDE_{indicator}_DEC {year}"
        buyer_col = f"BUY-SIDE_{indicator}_DEC {year}"

        # Check if columns exist
        if seller_col not in df.columns or buyer_col not in df.columns:
            print(f"Columns {seller_col} or {buyer_col} do not exist in dataset")
            continue

        # Extract relevant data
        df_indicator = df[[seller_col, buyer_col]].copy()

        # --- 1. Raw Data Correlations ---
        seller_vals = df_indicator[seller_col]
        buyer_vals = df_indicator[buyer_col]

        pearson_raw, pval_raw = pearsonr(seller_vals, buyer_vals)
        spearman_raw, sp_pval_raw = spearmanr(seller_vals, buyer_vals)

        print(f"\n=== {indicator} - {year} Raw Data Correlations ===")
        print(f"Pearson: {pearson_raw:.3f} (p-value: {pval_raw:.3g})")
        print(f"Spearman: {spearman_raw:.3f} (p-value: {sp_pval_raw:.3g})")

        # --- 2. Log-Transformed Data Correlations ---
        mask_positive = (seller_vals > 0) & (buyer_vals > 0)
        seller_log = np.log1p(seller_vals[mask_positive])
        buyer_log = np.log1p(buyer_vals[mask_positive])

        pearson_log, pval_log = pearsonr(seller_log, buyer_log)
        spearman_log, sp_pval_log = spearmanr(seller_log, buyer_log)

        print(f"\n=== {indicator} - {year} Log-Transformed Data Correlations ===")
        print(f"Pearson (log): {pearson_log:.3f} (p-value: {pval_log:.3g})")
        print(f"Spearman (log): {spearman_log:.3f} (p-value: {sp_pval_log:.3g})")

        # --- 3. LOWESS Smoothed Data Correlations ---
        lowess_frac = 0.5
        lowess_results = lowess(endog=buyer_vals, exog=seller_vals, frac=lowess_frac, return_sorted=True)
        seller_sorted = lowess_results[:, 0]
        buyer_lowess = lowess_results[:, 1]

        pearson_lowess, pval_lowess = pearsonr(seller_sorted, buyer_lowess)
        spearman_lowess, sp_pval_lowess = spearmanr(seller_sorted, buyer_lowess)

        print(f"\n=== {indicator} - {year} LOWESS Smoothed Data Correlations ===")
        print(f"Pearson (LOWESS): {pearson_lowess:.3f} (p-value: {pval_lowess:.3g})")
        print(f"Spearman (LOWESS): {spearman_lowess:.3f} (p-value: {sp_pval_lowess:.3g})")

        # --- 4. Box-Cox Transformation ---
        seller_positive = seller_vals[mask_positive]
        buyer_positive = buyer_vals[mask_positive]

        seller_boxcox, lambda_seller = boxcox(seller_positive)
        buyer_boxcox, lambda_buyer = boxcox(buyer_positive)

        pearson_boxcox, pval_boxcox = pearsonr(seller_boxcox, buyer_boxcox)
        spearman_boxcox, sp_pval_boxcox = spearmanr(seller_boxcox, buyer_boxcox)

        print(f"\n=== {indicator} - {year} Box-Cox Transformed Data Correlations ===")
        print(f"Pearson (Box-Cox): {pearson_boxcox:.3f} (p-value: {pval_boxcox:.3g})")
        print(f"Spearman (Box-Cox): {spearman_boxcox:.3f} (p-value: {sp_pval_boxcox:.3g})")

        # --- 5. Yeo-Johnson Transformation ---
        seller_array = seller_vals.values.reshape(-1, 1)
        buyer_array = buyer_vals.values.reshape(-1, 1)

        pt = PowerTransformer(method='yeo-johnson')
        seller_yj = pt.fit_transform(seller_array)
        buyer_yj = pt.fit_transform(buyer_array)

        pearson_yj, pval_yj = pearsonr(seller_yj.flatten(), buyer_yj.flatten())
        spearman_yj, sp_pval_yj = spearmanr(seller_yj.flatten(), buyer_yj.flatten())

        print(f"\n=== {indicator} - {year} Yeo-Johnson Transformed Data Correlations ===")
        print(f"Pearson (Yeo-Johnson): {pearson_yj:.3f} (p-value: {pval_yj:.3g})")
        print(f"Spearman (Yeo-Johnson): {spearman_yj:.3f} (p-value: {sp_pval_yj:.3g})")
```

```python
# Function to compute financial ratios only for SELL-SIDE to avoid data leakage
def compute_sell_side_financial_ratios(df):
    """
    Compute profitability ratios, leverage ratios, and cash flow ratios only for SELL-SIDE
    to prevent data leakage.
    """
    new_features = {}

    # Years available in the dataset
    years = ["2023", "2022", "2021", "2020", "2019"]

    for year in years:
        # Profitability Ratios
        new_features[f"SELL-SIDE_Gross_Margin_DEC {year}"] = df[f"SELL-SIDE_Gross Income_DEC {year}"] / (df[f"SELL-SIDE_Sales_DEC {year}"] + 1e-6)
        new_features[f"SELL-SIDE_EBIT_Margin_DEC {year}"] = df[f"SELL-SIDE_EBIT_DEC {year}"] / (df[f"SELL-SIDE_Sales_DEC {year}"] + 1e-6)
        new_features[f"SELL-SIDE_EBITDA_Margin_DEC {year}"] = df[f"SELL-SIDE_EBITDA_DEC {year}"] / (df[f"SELL-SIDE_Sales_DEC {year}"] + 1e-6)
        new_features[f"SELL-SIDE_Net_Profit_Margin_DEC {year}"] = df[f"SELL-SIDE_Net Income_DEC {year}"] / (df[f"SELL-SIDE_Sales_DEC {year}"] + 1e-6)

        # Leverage & Debt Ratios
        new_features[f"SELL-SIDE_Debt-to-Equity_Ratio_DEC {year}"] = df[f"SELL-SIDE_Total Debt_DEC {year}"] / (df[f"SELL-SIDE_Total Shareholders' Equity_DEC {year}"] + 1e-6)
        new_features[f"SELL-SIDE_Net_Debt-to-EBITDA_DEC {year}"] = df[f"SELL-SIDE_Net Debt_DEC {year}"] / (df[f"SELL-SIDE_EBITDA_DEC {year}"] + 1e-6)
        new_features[f"SELL-SIDE_Interest_Coverage_Ratio_DEC {year}"] = df[f"SELL-SIDE_EBIT_DEC {year}"] / (df[f"SELL-SIDE_Total Debt_DEC {year}"] + 1e-6)

        # Cash Flow Ratios
        new_features[f"SELL-SIDE_Cash_Flow_to_Debt_Ratio_DEC {year}"] = df[f"SELL-SIDE_Net Operating Cash Flow_DEC {year}"] / (df[f"SELL-SIDE_Total Debt_DEC {year}"] + 1e-6)
        new_features[f"SELL-SIDE_Free_Cash_Flow_Margin_DEC {year}"] = df[f"SELL-SIDE_Free Cash Flow_DEC {year}"] / (df[f"SELL-SIDE_Sales_DEC {year}"] + 1e-6)

    # Concatenate all new features at once to avoid fragmentation issues
    new_features_df = pd.DataFrame(new_features)

    # Merge efficiently
    df = pd.concat([df, new_features_df], axis=1)

    return df
```

```python
# Function to create statistical features
def create_statistical_features(df, base_features, prefix="SELL-SIDE"):
    """
    Create statistical summary features across base features, with a custom prefix.
    """
    df[f'{prefix}_feat_mean_base'] = df[base_features].mean(axis=1)
    df[f'{prefix}_feat_std_base'] = df[base_features].std(axis=1)
    df[f'{prefix}_feat_max_base'] = df[base_features].max(axis=1)
    df[f'{prefix}_feat_min_base'] = df[base_features].min(axis=1)
    df[f'{prefix}_feat_range_base'] = df[f'{prefix}_feat_max_base'] - df[f'{prefix}_feat_min_base']
    df[f'{prefix}_feat_sum_base'] = df[base_features].sum(axis=1)
    return df
```

```python
# Transformer Block for Financial Data
def transformer_block(inputs, num_heads=4, ff_dim=1600, dropout_rate=0.1):
    # Reshape input for Transformer (batch, sequence_length=1, features)
    x = Reshape((1, inputs.shape[-1]))(inputs)

    # Multi-head Self Attention
    attention_output = MultiHeadAttention(num_heads=num_heads, key_dim=inputs.shape[-1])(x, x)
    attention_output = Dropout(dropout_rate)(attention_output)
    attention_output = Add()([x, attention_output])
    attention_output = LayerNormalization(epsilon=1e-6)(attention_output)

    # Feed Forward Layer
    ff_output = Dense(ff_dim, activation="relu")(attention_output)
    ff_output = Dense(inputs.shape[-1], activation="linear")(ff_output)
    ff_output = Dropout(dropout_rate)(ff_output)
    ff_output = Add()([attention_output, ff_output])
    ff_output = LayerNormalization(epsilon=1e-6)(ff_output)

    # Reshape back to original dimensions (batch, features)
    return Reshape((inputs.shape[-1],))(ff_output)

# Build the Transformer Model
def build_transformer_model(input_dim, output_dim, num_heads=4, num_layers=2, ff_dim=1600, dropout_rate=0.1):
    inputs = Input(shape=(input_dim,))

    # Expand input for Transformer processing
    x = Dense(ff_dim, activation="relu")(inputs)

    # Stack multiple Transformer blocks
    for _ in range(num_layers):
        x = transformer_block(x, num_heads, ff_dim, dropout_rate)

    # Output layer for regression (no Softmax, use Linear activation)
    outputs = Dense(output_dim, activation="linear")(x)

    model = Model(inputs, outputs)
    return model
```

```python
import numpy as np
from sklearn.linear_model import Ridge

class MAMLRegression:
    def __init__(self, alpha=0.01, meta_lr=0.005, num_inner_steps=5, task_batch_size = 16):
        self.alpha = alpha  # Inner learning rate
        self.meta_lr = meta_lr  # Meta-learning rate
        self.num_inner_steps = num_inner_steps
        self.meta_model = Ridge(alpha=0.1)  # Base meta-model
        self.is_fitted = False  # Track whether the model is fitted

    def train(self, tasks, meta_epochs=100):
        for epoch in range(meta_epochs):
            meta_loss = 0.0
            meta_gradients = None

            for X_train, Y_train in tasks:
                # Convert to numpy arrays (ensure uniformity)
                X_train, Y_train = np.array(X_train), np.array(Y_train)

                # Ensure the meta model is fitted at least once
                if not self.is_fitted:
                    self.meta_model.fit(X_train, Y_train)
                    self.is_fitted = True

                # Clone meta model parameters for task adaptation
                task_model = Ridge(alpha=0.1)
                task_model.fit(X_train, Y_train)
                task_coefs = task_model.coef_.copy()

                # Inner loop gradient-based updates
                for _ in range(self.num_inner_steps):
                    preds = X_train @ task_coefs.T  # Prediction
                    gradients = 2 * X_train.T @ (preds - Y_train) / len(Y_train)  # Compute gradient
                    task_coefs -= self.alpha * gradients  # Update task-specific parameters

                # Compute task loss
                task_loss = np.mean((X_train @ task_coefs.T - Y_train) ** 2)
                meta_loss += task_loss

                # Accumulate gradients for meta-update
                if meta_gradients is None:
                    meta_gradients = gradients.copy()
                else:
                    meta_gradients += gradients

            # Apply meta-update if gradients exist
            if meta_gradients is not None:
                current_params = self.meta_model.coef_ if hasattr(self.meta_model, "coef_") else np.zeros(X_train.shape[1])
                meta_update = current_params - self.meta_lr * (meta_gradients / len(tasks))

                # Re-train meta model with updated coefficients
                self.meta_model.coef_ = meta_update  # Manually update weights

            if epoch % 10 == 0 or epoch == meta_epochs - 1:
                print(f"Epoch {epoch}, Meta Loss: {meta_loss / len(tasks)}")

    def predict(self, X):
        if not self.is_fitted:
            raise ValueError("Model has not been trained yet. Please train before predicting.")
        return X @ self.meta_model.coef_.T  # Direct matrix multiplication

# Example usage:
maml_regressor = MAMLRegression()
maml_regressor.train(train_tasks)
```

```python
import numpy as np
from sklearn.linear_model import BayesianRidge
from sklearn.multioutput import MultiOutputRegressor
from copy import deepcopy

class MAMLBayesianRegression:
    def __init__(self, alpha=0.001, meta_lr=0.0001, num_inner_steps=5, 
                 alpha_1=1, alpha_2=1, lambda_1=100, lambda_2=100, tol=1e-2):
        self.alpha = alpha  # Inner learning rate for task adaptation
        self.meta_lr = meta_lr  # Meta-learning rate
        self.num_inner_steps = num_inner_steps
        self.meta_model = MultiOutputRegressor(BayesianRidge(alpha_1=alpha_1, alpha_2=alpha_2, 
                                                             lambda_1=lambda_1, lambda_2=lambda_2, tol=tol))
        self.fitted = False  # Track if the meta-model has been initialized

    def train(self, tasks, meta_epochs=10):
        # Fit the meta-model on the first task to initialize parameters
        X_sample, Y_sample = tasks[0]  
        self.meta_model.fit(X_sample, Y_sample)
        self.fitted = True  

        for epoch in range(meta_epochs):
            meta_loss = 0.0
            num_targets = Y_sample.shape[1]  # Number of output variables
            num_features = X_sample.shape[1]  # Number of input features

            meta_gradients = np.zeros((num_targets, num_features))  # Initialize accumulated gradients

            for X_train, Y_train in tasks:
                # Clone the meta model for task adaptation
                task_model = deepcopy(self.meta_model)
                task_model.fit(X_train, Y_train)  # First fit

                # Inner loop adaptation (fine-tuning)
                for _ in range(self.num_inner_steps):
                    task_model.fit(X_train, Y_train)

                # Compute task-specific loss
                task_predictions = task_model.predict(X_train)
                task_loss = np.mean((task_predictions - Y_train) ** 2)
                meta_loss += task_loss

                # Compute difference between adapted and original meta-model
                task_gradients = np.array([task_model.estimators_[i].coef_ - self.meta_model.estimators_[i].coef_ 
                                           for i in range(num_targets)])

                # Accumulate gradients for meta-update (element-wise addition)
                meta_gradients += task_gradients

            # Meta-update (apply weighted update)
            for i in range(num_targets):
                self.meta_model.estimators_[i].coef_ += self.meta_lr * (meta_gradients[i] / len(tasks))

            # Print progress
            if epoch % 10 == 0 or epoch == meta_epochs - 1:
                print(f"Epoch {epoch}, Meta Loss: {meta_loss / len(tasks)}")

    def predict(self, X):
        if not self.fitted:
            raise ValueError("Meta model has not been trained yet.")
        return self.meta_model.predict(X)

# Initialize and train MAML Bayesian Regression model (Multi-Output)
maml_bayesian = MAMLBayesianRegression(alpha=0.001, meta_lr=0.0001, num_inner_steps=5,
                                       alpha_1=1, alpha_2=1, lambda_1=100, lambda_2=100, tol=1e-2)
maml_bayesian.train(train_tasks)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

class MAMLRidgeRegression:
    def __init__(self, input_dim, alpha=0.01, meta_lr=0.001, num_inner_steps=3, ridge_lambda=1.0):
        self.alpha = alpha  # Inner loop learning rate
        self.meta_lr = meta_lr  # Meta-learning rate
        self.num_inner_steps = num_inner_steps
        self.ridge_lambda = ridge_lambda  # Regularization strength

        # Ridge regression model
        self.input_dim = input_dim
        self.meta_model = nn.Linear(input_dim, input_dim, bias=True)  # Bias enabled for Ridge regression
        self.meta_optimizer = optim.Adam(self.meta_model.parameters(), lr=meta_lr)

    def ridge_loss(self, y_pred, y_true, model):
        """Custom Ridge Regression Loss = MSE + L2 Regularization"""
        mse_loss = nn.MSELoss()(y_pred, y_true)
        l2_reg = sum(torch.norm(param) ** 2 for param in model.parameters())  # L2 Regularization
        return mse_loss + self.ridge_lambda * l2_reg  # Ridge loss

    def inner_loop(self, X, y):
        """Performs task-specific adaptation (inner loop)"""
        adapted_model = nn.Linear(self.input_dim, self.input_dim, bias=True)
        adapted_model.load_state_dict(self.meta_model.state_dict())  # Clone meta-model weights

        optimizer = optim.SGD(adapted_model.parameters(), lr=self.alpha)

        for _ in range(self.num_inner_steps):
            y_pred = adapted_model(X)
            loss = self.ridge_loss(y_pred, y, adapted_model)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        return adapted_model

    def train(self, tasks, meta_epochs=100):
        for epoch in range(meta_epochs):
            meta_loss = 0.0
            self.meta_optimizer.zero_grad()  # Reset meta optimizer gradients

            for X_train, Y_train in tasks:
                X_train = torch.tensor(X_train, dtype=torch.float32)
                Y_train = torch.tensor(Y_train, dtype=torch.float32)

                # Ensure all tensors have the correct shape
                if X_train.shape[1] != self.input_dim:
                    print(f"Skipping task with shape mismatch: {X_train.shape[1]} features (expected {self.input_dim})")
                    continue

                adapted_model = self.inner_loop(X_train, Y_train)  # Perform inner-loop updates

                # Compute second-order gradients for meta-update
                y_pred = adapted_model(X_train)
                loss = self.ridge_loss(y_pred, Y_train, adapted_model)

                # Compute gradients w.r.t. the initial meta-model parameters
                grads = torch.autograd.grad(loss, adapted_model.parameters(), retain_graph=True, create_graph=True)

                # Accumulate gradients for meta-update
                for meta_param, grad in zip(self.meta_model.parameters(), grads):
                    if meta_param.grad is None:
                        meta_param.grad = grad.clone()
                    else:
                        meta_param.grad += grad

                meta_loss += loss.item()

            self.meta_optimizer.step()  # Meta-update step

            if epoch % 10 == 0 or epoch == meta_epochs - 1:
                print(f"Epoch {epoch}, Meta Loss: {meta_loss / len(tasks)}")

    def predict(self, X):
        """Predict using the trained MAML model"""
        X = torch.tensor(X, dtype=torch.float32)
        return self.meta_model(X).detach().numpy()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np

# ---------------------- Memory-Augmented Meta-Learner ----------------------
class MANN(Model):
    def __init__(self, input_dim, memory_size=200, embedding_dim=128):
        super(MANN, self).__init__()
        self.input_dim = input_dim
        self.memory_size = memory_size
        self.embedding_dim = embedding_dim

        # Memory Module (Key-Value Store)
        self.memory_keys = tf.Variable(tf.random.normal([memory_size, input_dim]), trainable=False)
        self.memory_values = tf.Variable(tf.random.normal([memory_size, input_dim]), trainable=False)

        # Controller (LSTM for adaptive learning)
        self.controller = layers.LSTM(256, return_sequences=False, return_state=True)

        # Fully connected network for predictions
        self.fc = layers.Dense(input_dim)

    def retrieve_memory(self, x):
        """Retrieve relevant past knowledge from memory"""
        similarities = tf.matmul(x, self.memory_keys, transpose_b=True)  # Compute similarity
        weights = tf.nn.softmax(similarities * 5, axis=-1)  # Softmax attention weights
        retrieved_info = tf.matmul(weights, self.memory_values)  # Weighted sum of memory values
        return retrieved_info

    def update_memory(self, x, y):
        """Update memory dynamically based on batch size"""
        batch_size = tf.shape(x)[0]  # Get current batch size dynamically
        indices = tf.random.uniform([batch_size], maxval=self.memory_size, dtype=tf.int32)  # Match batch size

        # Ensure updates fit within available memory slots
        self.memory_keys.assign(tf.tensor_scatter_nd_update(self.memory_keys, indices[:, None], x))
        self.memory_values.assign(tf.tensor_scatter_nd_update(self.memory_values, indices[:, None], y))


    def call(self, x, training=True):
        """Forward pass: Combine retrieved memory and new task info"""
        retrieved_info = self.retrieve_memory(x)
        x_combined = tf.concat([x, retrieved_info], axis=-1)

        # Pass through LSTM Controller
        lstm_out, _, _ = self.controller(tf.expand_dims(x_combined, axis=1))
        x_final = self.fc(lstm_out)

        if training:
            self.update_memory(x, x_final)  # Store new task knowledge
        return x_final

# ---------------------- Training Process ----------------------
def train_mann(model, train_tasks, val_tasks, epochs=100, batch_size=16, lr=0.001):
    optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
    loss_fn = tf.keras.losses.MeanSquaredError()

    for epoch in range(epochs):
        total_loss = 0.0
        for X_train, Y_train in train_tasks:
            with tf.GradientTape() as tape:
                predictions = model(X_train, training=True)
                loss = loss_fn(Y_train, predictions)
            
            grads = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(grads, model.trainable_variables))
            total_loss += loss.numpy()

        if epoch % 10 == 0 or epoch == epochs - 1:
            print(f"Epoch {epoch}, Loss: {total_loss / len(train_tasks)}")

# ---------------------- MANN Training Execution ----------------------
input_dim = train_tasks[0][0].shape[1]  # Extract feature size
mann_model = MANN(input_dim)

# Train the model
train_mann(mann_model, train_tasks, val_tasks, epochs=100)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.layers import LeakyReLU

class MANN_fnn(Model):
    def __init__(self, input_dim, output_dim, memory_size=400, embedding_dim=128):
        super(MANN_fnn, self).__init__()
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.memory_size = memory_size
        self.embedding_dim = embedding_dim

        # Memory Module (Key-Value Store)
        self.memory_keys = tf.Variable(tf.random.normal([memory_size, input_dim]), trainable=False)
        self.memory_values = tf.Variable(tf.random.normal([memory_size, output_dim]), trainable=False)  # ✅ FIXED

        # Controller (Feedforward Neural Network)
        self.controller = tf.keras.Sequential([
            layers.Dense(64),
            LeakyReLU(alpha=0.2),
            layers.Dense(32),
            LeakyReLU(alpha=0.2)
        ])

        # Output layer that matches output_dim
        self.fc = layers.Dense(output_dim)  # ✅ FIXED

    def retrieve_memory(self, x):
        """Retrieve relevant past knowledge from memory"""
        similarities = tf.matmul(x, self.memory_keys, transpose_b=True)  # Compute similarity
        weights = tf.nn.softmax(similarities * 5, axis=-1)  # Sharpened Softmax attention
        retrieved_info = tf.matmul(weights, self.memory_values)  # Weighted sum of memory values
        return retrieved_info

    def update_memory(self, x, y):
        """Update memory dynamically based on batch size"""
        batch_size = tf.shape(x)[0]  # Get current batch size dynamically
        indices = tf.random.uniform([batch_size], maxval=self.memory_size, dtype=tf.int32)  # Match batch size

        # ✅ FIXED: Ensure updates match memory storage dimensions
        self.memory_keys.assign(tf.tensor_scatter_nd_update(self.memory_keys, indices[:, None], x))
        self.memory_values.assign(tf.tensor_scatter_nd_update(self.memory_values, indices[:, None], y))  # ✅ FIXED

    def call(self, x, training=True):
        """Forward pass: Combine retrieved memory and new task info"""
        retrieved_info = self.retrieve_memory(x)
        x_combined = tf.concat([x, retrieved_info], axis=-1)

        # Pass through FNN
        x_final = self.controller(x_combined)
        x_final = self.fc(x_final)

        if training:
            self.update_memory(x, x_final)  # ✅ FIXED: Update memory with correct shape
        return x_final
```

```python
# ---------------------- Training Process ----------------------
def train_mann_fnn(model, train_tasks, val_tasks, epochs=10, batch_size=128, lr=0.0001):
    optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
    loss_fn = tf.keras.losses.MeanSquaredError()

    for epoch in range(epochs):
        total_loss = 0.0
        for X_train, Y_train in train_tasks:
            with tf.GradientTape() as tape:
                predictions = model(X_train, training=True)
                loss = loss_fn(Y_train, predictions)
            
            grads = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(grads, model.trainable_variables))
            total_loss += loss.numpy()

        if epoch % 10 == 0 or epoch == epochs - 1:
            print(f"Epoch {epoch}, Loss: {total_loss / len(train_tasks)}")

# ---------------------- MANN Training Execution ----------------------
input_dim = train_tasks[0][0].shape[1]  # Extract feature size
output_dim = train_tasks[0][1].shape[1]
mann_model_fnn = MANN_fnn(input_dim, output_dim)    

# Train the model
train_mann_fnn(mann_model_fnn, train_tasks, val_tasks, epochs=100)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

def compute_financial_ratios(df):
    """
    Compute sophisticated financial ratios given a dataframe with sell-side and buy-side financial indicators.
    
    Args:
        df (pd.DataFrame): DataFrame containing financial metrics.

    Returns:
        pd.DataFrame: DataFrame with additional financial ratio features.
    """
    df = df.copy()  # Ensure we don't modify the original DataFrame
    
    # Ensure column names are unique
    df = df.loc[:, ~df.columns.duplicated()]
    
    # Avoid division errors by replacing zero or missing values with a very small value
    epsilon = 1e-6

    # Ensure relevant columns exist in the DataFrame before computing ratios
    required_columns = [
        "Total Liabilities", "Total Shareholders' Equity", "EBITDA",
        "Net Debt", "Total Assets", "Total Debt", "Cash & Short-Term Investments"
    ]
    missing_columns = [col for col in required_columns if col not in df.columns]
    
    if missing_columns:
        raise ValueError(f"Missing required columns for ratio calculations: {missing_columns}")
    
    # Compute financial ratios safely
    df["Debt-to-Equity Ratio"] = df["Total Liabilities"] / (df["Total Shareholders' Equity"].replace(0, epsilon))
    df["EBITDA Margin"] = df["EBITDA"] / (df["Sales"].replace(0, epsilon))
    df["Return on Equity (ROE)"] = df["Net Income"] / (df["Total Shareholders' Equity"].replace(0, epsilon))
    df["Return on Assets (ROA)"] = df["Net Income"] / (df["Total Assets"].replace(0, epsilon))
    df["Debt-to-EBITDA Ratio"] = df["Total Debt"] / (df["EBITDA"].replace(0, epsilon))
    df["Net Profit Margin"] = df["Net Income"] / (df["Sales"].replace(0, epsilon))
    df["Asset Turnover"] = df["Sales"] / (df["Total Assets"].replace(0, epsilon))

    return df
```
